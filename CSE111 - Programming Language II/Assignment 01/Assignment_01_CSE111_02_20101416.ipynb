```
#Easy: (1)
div_cnt = 0
for num in range(3, 31):
  if num%3 == 0 or num%5 == 0 or num%15 == 0:
    div_cnt += 1
    print(num)
print('Total numbers:',div_cnt)
```

```
#Easy: (2)
a, b, c = input().split(' ')
if b <= a >= c:
  print(a)
elif a <= b >= c:
  print(b)
elif a <= c >= b:
  print(c)
```

```
#Easy: (3)
marks = int(input())
if marks >= 90:
  print('A')
elif marks >= 85:
  print('A-')
elif marks >= 80:
  print('B+')
elif marks >= 75:
  print('B')
elif marks >= 70:
  print('B-')
elif marks >= 65:
  print('C+')
elif marks >= 60:
  print('C')
elif marks >= 57:
  print('C-')
elif marks >= 55:
  print('D+')
elif marks >= 52:
  print('D')
elif marks >= 50:
  print('D-')
else:
  print('F')
```

```
#Easy: (3)_V2
grade_marks = '90 85 80 75 70 65 60 57 55 52 50'.split(' ')
grade_value = 'A A- B+ B B- C+ C C- D+ D D- F'.split(' ')
marks = int(input())
for i in range(len(grade_marks)):
  if marks >= int(grade_marks[i]):
    print(grade_value[i])
    break
else:
  print(grade_value[-1])
```

```
#Easy: (4)
n = int(input())
if n<2:
    print('Not Prime')
else:
  for i in range(2,n):
    if n%i==0:
      print('Not Prime')
      break
  else:
    print('Prime')
```

```
#Medium: (1)
n, a, b = int(input()), 0, 1
print(a, end=' ')
while b < n:
  print(b, end=' ')
  a, b = b, a+b
```

```
#Medium: (2)
n = int(input())
r = 0
while n != 0:
  r = r*10 + n%10
  n = n//10
print(r)
```

```
#Medium: (3)
n, s = int(input()), 0
for i in range(1,n):
  if n%i == 0:
    s += i
if s == n:
  print('Perfect Number')
else:
  print('Not Perfect')
```

```
#Hard: (1)
l, n = [], int(input())
while not n == 0:
  i = n%10
  n = n//10
  if not i in l:
    l.append(i)
print(len(l),'.',sep='')
```

```
#Hard: (2)
n = int(input())
#upper tip
print(' '*(n-1) + '*')
#upper half
for i in range(n-2, 0, -1):
  print(' '*i + '*' + ' '*(2*n-2*i-3) + '*')
#middle line
print('*'*(2*n-1))
#lower half
for i in range(1, n-1):
  print(' '*i + '*' + ' '*(2*n-2*i-3) + '*')
#lower tip
print(' '*(n-1) + '*')
```
