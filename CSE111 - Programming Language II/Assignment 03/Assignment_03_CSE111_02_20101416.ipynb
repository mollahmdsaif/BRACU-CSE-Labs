```
#Easy: (2)
def is_palindrome(string):
  for i in range(len(string)//2):
    if string[i] != string[-i-1]:
      return False
  else:
    return True

input_str = input()
if is_palindrome(input_str):
  print('Palindrome')
else:
  print('Not Palindrome')
```

```
#Medium: (1)
input_str = input()
lower_case_cnt = 0
upper_case_cnt = 0
for c in input_str:
  if 'a' <= c <= 'z':
    lower_case_cnt += 1
  elif 'A' <= c <= 'Z':
    upper_case_cnt += 1
if lower_case_cnt < upper_case_cnt:
  print(input_str.upper())
else:
  print(input_str.lower())
```

```
#Medium: (1)_V2
def index_of(string, chr):
  for i in range(len(string)):
    if chr == string[i]:
      return i
  return -1

def to_lower_case(string):
  return_str = ''
  lower_case = 'abcdefghijklmnopqrstuvwxyz'
  upper_case = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
  for c in string:
    if 'A' <= c <= 'Z':
      return_str = return_str + lower_case[index_of(upper_case, c)]
    else:
      return_str = return_str + c
  return return_str

def to_upper_case(string):
  return_str = ''
  lower_case = 'abcdefghijklmnopqrstuvwxyz'
  upper_case = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
  for c in string:
    if 'a' <= c <= 'z':
      return_str = return_str + upper_case[index_of(lower_case, c)]
    else:
      return_str = return_str + c
  return return_str

input_str = input()
lower_case_cnt = 0
upper_case_cnt = 0
for c in input_str:
  if 'a' <= c <= 'z':
    lower_case_cnt += 1
  elif 'A' <= c <= 'Z':
    upper_case_cnt += 1
if lower_case_cnt < upper_case_cnt:
  print(to_upper_case(input_str))
else:
  print(to_lower_case(input_str))
```

```
#Medium: (2)
input_str = input()
num_flag = False
word_flag = False
for c in input_str:
  if not word_flag and ('a' <= c <= 'z' or 'A' <= c <= 'Z'):
    word_flag = True
  elif not num_flag and ('0' <= c <= '9'):
     num_flag = True
  if num_flag and word_flag:
    print('MIXED')
    break
else:
  if word_flag:
    print('WORD')
  else:
    print('NUMBER')
```

```
#Medium: (3)
def find_upper_case(string, starting = 0):
  for i in range(starting, len(string)):
    if 'A' <= string[i] <= 'Z':
      return i

input_str = input()
starting_index = find_upper_case(input_str)
ending_index = find_upper_case(input_str, starting_index+1)
if ending_index - starting_index == 1:
  print('BLANK')
else:
  print(input_str[starting_index+1:ending_index])
```

```
#Medium: (4)
def index_of(string, substring):
  for i in range(len(string)-7):
    if string[i:i+8] == substring:
      return i

input_str = input()
if 'too good' in input_str:
  i = index_of(input_str, 'too good')
  print(input_str[:i:] + 'excellent' + input_str[i+8::])
else:
  print(input_str)
```

```
#Hard: (1)
l = input().split(', ')
output_str = ''
for chr in l[0]:
  if chr in l[1]:
    output_str += chr
for chr in l[1]:
  if chr in l[0]:
    output_str += chr
if output_str == '':
  print('Nothing in common')
else:
  print(output_str)
```

```
#Hard: (3)
input_str = input()
lower_flag = upper_flag = digit_flag = sp_flag = False
for chr in input_str:
  if 'a' <= chr <= 'z':
    lower_flag = True
  elif 'A' <= chr <= 'Z':
    upper_flag = True
  elif '0' <= chr <= '9':
    digit_flag = True
  else:
    sp_flag = True
  if lower_flag and upper_flag and digit_flag and sp_flag:
    break
output_str = ''
output_str += '' if lower_flag else 'Lowercase missing, '
output_str += '' if upper_flag else 'Uppercase missing, '
output_str += '' if digit_flag else 'Digit missing, '
output_str += '' if sp_flag else 'Special character missing, '
output_str += 'OK, ' if output_str == '' else ''
print(output_str[:-2])
```
