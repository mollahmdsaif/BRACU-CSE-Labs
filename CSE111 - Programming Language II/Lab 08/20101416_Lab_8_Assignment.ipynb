A vehicle assumes that the whole world is a 2-dimensional graph paper. It maintains its x and y coordinates (both are integers). The vehicle gets manufactured (constructed) at (0, 0) coordinate.

#### Subtasks:
1. Design a <b>Vehicle2010 class</b> which inherits movement methods from <b>Vehicle</b> and adds new methods called <b>move UpperRight, UpperLeft, LowerRight, LowerLeft.</b> Each of these diagonal move methods must re-use two inherited and appropriate move methods. Write a user class as well which will show that all of your methods are working.
2. Write an <b>“equals”</b> method which tests if significant class properties are the same (in this case x and y).

<b>Note: All moves are 1 step. That means a single call to any move method changes value of either x or y or both by 1.</b>

```python
class Vehicle:
    def __init__(self):
        self.x = 0
        self.y = 0
    def moveUp(self):
        self.y+=1
    def moveDown(self):
        self.y-=1
    def moveRight(self):
        self.x+=1
    def moveLeft(self):
        self.x-=1
    def __str__(self):
        return '('+str(self.x)+' , '+str(self.y)+')'


class Vehicle2010(Vehicle):
    def __init__(self):
        super().__init__()
    
    def moveUpperRight(self):
        self.moveUp()
        self.moveRight()
        
    def moveUpperLeft(self):
        self.moveUp()
        self.moveLeft()
        
    def moveLowerRight(self):
        self.moveDown()
        self.moveRight()
        
    def moveLowerLeft(self):
        self.moveDown()
        self.moveLeft()
        
    def equals(self, other):
        return self.x == other.x and self.y == other.y
    

def main():
    print('Part 1')
    print('------')
    car = Vehicle()
    print(car)
    car.moveUp()
    print(car)
    car.moveLeft()
    print(car)
    car.moveDown()
    print(car)
    car.moveRight()
    print(car)
    print('------')
    print('Part 2')
    print('------')
    car1 = Vehicle2010()
    print(car1)
    car1.moveLowerLeft()
    print(car1)
    car2 = Vehicle2010()
    car2.moveLeft()
    print(car1.equals(car2))
    car2.moveDown()
    print(car1.equals(car2))
    print('------')

if __name__=='__main__':
    main()
```

```python
class RealNumber:
    def __init__(self, r=0):
        self.__realValue = r
    def getRealValue(self):
        return self.__realValue
    def setRealValue(self, r):
        self.__realValue = r
    def ping(self):
        print('I am in RealNumber class')
    def __str__(self):
        return 'RealPart: '+str(self.getRealValue())
    
class ComplexNumber(RealNumber):
    def __init__(self, *info):
        if len(info) != 0:
            super().__init__(info[0])
            self.__imaginaryValue = info[1]
        else:
            super().__init__(1.0)
            self.__imaginaryValue = 1.0
            
    def getImaginaryValue(self):
        return self.__imaginaryValue
    
    def setImaginaryValue(self, r):
        self.__imaginaryValue = r
        
    def ping(self):
        print('I am in ImaginaryNumber class')
            
    def __str__(self):
        return f'RealPart: {self.getRealValue():.1f}\nImaginaryPart: {self.__imaginaryValue:.1f}'
    
    

def main():
    cn1 = ComplexNumber()
    print(cn1)
    print('---------')
    
    cn2 = ComplexNumber(5,7)
    print(cn2)
    print('---------')

if __name__=='__main__':
    main()
```

```python
class Fruit:
    def __init__(self, formalin=False, name=''):
        self.__formalin = formalin
        self.name = name
    def getName(self):
        return self.name
    def hasFormalin(self):
        return self.__formalin

class testFruit:
    def test(self, f):
        print('----Printing Detail----')
        if f.hasFormalin():
            print('Do not eat the ',f.getName(),'.',sep='')
            print(f)
        else:
            print('Eat the ',f.getName(),'.',sep='')
            print(f)
            
class Mango(Fruit):
    def __init__(self):
        super().__init__(formalin=True, name='Mango')
        
    def __str__(self):
        return 'Mangos are bad for you'
        
class Jackfruit(Fruit):
    def __init__(self):
        super().__init__(name='Jackfruit')
        
    def __str__(self):
        return 'Jackfruits are good for you'


def main():
    m = Mango()
    j = Jackfruit()
    t1 = testFruit()
    t1.test(m)
    t1.test(j)

if __name__=='__main__':
    main()
```
