```python
class Student:
    ID = 0
    def __init__(self, name, department, age, cgpa):
        Student.ID += 1
        self.ID = Student.ID
        self.name = name
        self.department = department
        self.age = age
        self.cgpa = cgpa

    def get_details(self):
        print('ID:', self.ID)
        print('Name:', self.name)
        print('Department:', self.department)
        print('Age:', self.age)
        print('CGPA:', self.cgpa)

    @classmethod
    def from_String(cls, info):
        name, department, age, cgpa = info.split('-')
        temp_obj = cls(name, department, age, cgpa)
        return temp_obj


def main():
    s1 = Student("Samin", "CSE", 21, 3.91)
    s1.get_details()
    print("-----------------------")
    s2 = Student("Fahim", "ECE", 21, 3.85)
    s2.get_details()
    print("-----------------------")
    s3 = Student("Tahura", "EEE", 22, 3.01)
    s3.get_details()
    print("-----------------------")
    s4 = Student.from_String("Sumaiya-BBA-23-3.96")
    s4.get_details()


if __name__=='__main__':
    main()
```

```python
class Dates:
    def __init__(self, date_str):
        day, month, year = date_str.split('-')
        self.day = int(day)
        self.month = int(month)
        self.year = int(year)


    @classmethod
    def toDashDate(cls, date_str):
        temp_obj = cls(date_str.replace('/', '-'))
        return temp_obj

    def getDate(self):
        return [self.day, self.month, self.year]


'''
*Subtask 4: In the conditional statement it prints "Equal". Explain why.
*Ans- In case of date1, we get Dates object with 'day', 'month' and 'year' data. On the
      other hand, in case date2, the string of date have delimiter '/'.
      But since our constructor only receives string of date with delimiter '-', we
      have to make a class method 'toDashDate()' which accepts string of date with 
      delimiter '/' and replaces it to '-' for date2. Then we pass it to class constructor.
      So, for both case the arguments of the constructors are same.
      This is why when we compare getDate for both dates, it prints "Equal".

'''


def main():
    date1 = Dates("05-09-2020")
    dateFromDB = "05/09/2020"
    date2= Dates.toDashDate(dateFromDB)
    if(date1.getDate() == date2.getDate() ):
        print("Equal")
    else:
        print("Unequal")


if __name__=='__main__':
    main()
```

```python
from math import pi as PI

class Circle:
    def __init__(self, radius):
        self.__radius = radius

    def area(self):
        return PI  * self.__radius * self.__radius

    def getRadius(self):
        return self.__radius

    def setRadius(self, radius):
        self.__radius = radius

    def __add__(self, other):
        radius_sum = self.__radius + other.__radius
        temp_circle_obj = Circle(radius_sum)
        return temp_circle_obj


def main():
    c1 = Circle(4)
    print("First circle radius:" , c1.getRadius())
    print("First circle area:" ,c1.area())

    c2 = Circle(5)
    print("Second circle radius:" ,c2.getRadius())
    print("Second circle area:" ,c2.area())

    c3 = c1 + c2
    print("Third circle radius:" ,c3.getRadius())
    print("Third circle area:" ,c3.area())


if __name__=='__main__':
    main()
```

#### Subtasks:
1. Create a Doll class.
2. Create the required constructor.
3. Write a method to print the name and the price of the object
4. Use operator overloading for the addition operators.
5. Write a method to handle operator overloading for the “>” logical operator that compares the price of the objects.

##### <i>Hints:</i>
- <i>Notice that the price of each object is being checked with the price of obj in the given code.</i>
- <i>Notice the word Doll in the first 4 outputs and the last output. You have to print exactly as represented here.</i>

```python
class Dolls:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def detail(self):
        return f"Doll: {self.name}\nTotal Price: {self.price} taka"

    def __add__(self, other):
        added_name = self.name + ' ' + other.name
        added_price = self.price + other.price
        temp_doll_obj = Dolls(added_name, added_price)
        return temp_doll_obj

    def __gt__(self, other):
        return self.price > other.price


def main():
    obj_1 = Dolls("Tweety", 2500)
    print(obj_1.detail())
    if obj_1 > obj_1:
        print("Congratulations! You get the Tweety as a gift!")
    else:
        print("Thank you!")

    print("=========================")
    obj_2 = Dolls("Daffy Duck", 1800)
    print(obj_2.detail())
    if obj_2 > obj_1:
        print("Congratulations! You get the Tweety as a gift!")
    else:
        print("Thank you!")

    print("=========================")
    obj_3 = Dolls("Bugs Bunny", 3000)
    print(obj_3.detail())
    if obj_3 > obj_1:
        print("Congratulations! You get the Tweety as a gift!")
    else:
        print("Thank you!")

    print("=========================")
    obj_4 = Dolls("Porky Pig", 1500)
    print(obj_4.detail())
    if obj_4 > obj_1:
        print("Congratulations! You get the Tweety as a gift!")
    else:
        print("Thank you!")

    print("=========================")
    obj_5 = obj_2 + obj_3
    print(obj_5.detail())
    if obj_5 > obj_1:
        print("Congratulations! You get the Tweety as a gift!")
    else:
        print("Thank you!")


if __name__=='__main__':
    main()
```
