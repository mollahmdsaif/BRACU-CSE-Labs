```
#Easy (1)
def fnc(a, b):
  if a*b == 0 or a%b == 0:
    return 0
  return a/b - a//b
print(fnc(9, 4))
print(fnc(5, 0))
print(fnc(0, 5))
```

```
#Easy (2)
def bmi(h, w):
  bmi_value = w / ((h/100)**2)
  r = 'Score is {:.1f}.You are '.format(bmi_value)
  if bmi_value < 18.5:
    r += 'Underweight'
  elif 18.5 <= bmi_value <= 24.9:
    r += 'Normal'
  elif 25 <= bmi_value <= 30:
    r += 'Overweight'
  else:
    r += 'Obese'
  print(r)

bmi(152,48)
```

```
#Easy (3)
def fnc(low, high, divi):
  s = 0
  for i in range(low, high):
    if i % divi == 0:
      s += i
  return s

print(fnc(0,10,2))
```

```
#Medium (1)
def fnc(*order):
  menu = ['BBQ Chicken Cheese Burger', 'Beef Burger', 'Naga Drums']
  price = [270, 183.6, 216]
  t = 40
  for i in range(len(order)):
    if order[i] in menu:
      t += price[menu.index(order[i])]
    else:
      t += 20
  return t

print(fnc('Beef Burger'))
```

```
#Medium (2)
def replace_domain(email, *domain):
  r = 'Changed: ' if len(domain)!=1 else 'Unchanged: '
  for c in email:
    if c != '@':
      r += c
    else:
      return r + '@' + domain[0]

print(replace_domain('bob@sheba.xyz', 'sheba.xyz'))
```

```
#Medium (3)
def find_vowels(s):
  vowels, r = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U'], ''
  for c in s:
    if c in vowels:
      r += (c + ', ')
  return 'No vowels in the name' if r=='' else 'Vowels: ' + r[:-2] + '. Total number of vowels: ' + str(len(r)//3)
print(find_vowels('XYZ'))
```

```
#Medium (4)
def is_palindrome(s):
  s = s.replace(' ', '')
  for i in range(len(s)//2):
    if not s[i] == s[-i-1]:
      return 'Not a palindrome'
  else:
    return 'Palindrome'

print(is_palindrome('nurses run'))
```

```
#Hard (1)
def has_passed_years(date1, years):
  from datetime import datetime
  date2 = str(datetime.now())[:10]
  if int(date2[:4]) - int(date1[:4]) < years:
    return False
  elif int(date2[:4]) - int(date1[:4]) == years:
    if date2[5:7] < date1[5:7]:
      return False
    elif date2[5:7] == date1[5:7]:
      if int(date2[8:]) >= int(date1[8:]):
        return True
      else:
        return False
    else:
      return True
  else:
    return True

def salary_increase(*info):
  for i in range(0,len(info), 3):
    if has_passed_years(info[i+2], 10):
      print(info[i] + ': ' + str(info[i+1]//20*3+15000))
    elif has_passed_years(info[i+2], 5):
      print(info[i] + ': ' + str(info[i+1]//10+5000))
    else:
      print(info[i] + ': ' + str(info[i+1]//10))

salary_increase('Alice', 20000, '2017-03-21', 'Bob', 50000, '2007-05-10')
```

```
#Hard (2)
def fnc(n):
  l = lambda x: '' if x==1 else 's'
  y = n//365
  m = (n - y*365)//30
  d = (n - y*365 - m*30)
  return str(y) + ' year' + l(y) + ', ' +  str(m) + ' month' + l(m) + ', ' + str(d) + ' day' + l(d)

print(fnc(4316))
```

```
#Hard (3)
def capitalize(s):
  l = s.capitalize().replace(' i ', ' I ').split(' ')
  for i in range(1,len(l)-1):
    if l[i-1][-1] in ['.', '!', '?']:
      l[i] = l[i].capitalize()
  return ' '.join(l)

s = 'my favourite animal is a dog. a dog has sharp teeth so that it can eat flesh very easily. do you know my pet dog’s name? i love my pet very much.'
print(capitalize(s))
```
